```python
"""
Programa soma
Descrição: Esse programa lê dois números do teclado, calcula a soma e coloca o resultado na tela.
Autor: Natália
Data: 24/03/2026
Versão: 0.0.1
"""
```

```python
# Alocação de memória

primeira_parcela = 0

segunda_parcela = 0

soma = 0
```

```python
# Entrada de dados
primeira_parcela = int(input("primeira_parcela"))

segunda_parcela = int(input("segunda_parcela"))
```

```python
# Processamento de dados
soma = primeira_parcela + segunda_parcela
```

```python
# Saída de dados
print(soma)
```
